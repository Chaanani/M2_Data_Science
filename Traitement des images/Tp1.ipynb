```python
import pandas as pd
import numpy as np
```

```python
data5 =pd.read_csv("Summary of 58-1.csv")
data5['label']= [5 for i in range(len(data5))]

data9 =pd.read_csv("Summary of LED-1.csv")

data9['label']= [9 for i in range(len(data9))]
```

```python
Data = pd.concat([data5, data9])
```

```python
Data
```

```python
y_train = Data["label"]
x_train = Data.drop("label", axis=1)
```

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt

correlation_matrix = Data.corr()
feature_correlation = correlation_matrix['label'].abs().sort_values(ascending=False)
feature_correlation = feature_correlation.drop(labels = ['label','Slice'])
plt.figure(figsize=(10, 6))
feature_correlation.plot(kind='bar', color='skyblue')
plt.title('Histogram of Feature Correlations with Target')
plt.xlabel('Features')
plt.ylabel('Absolute Correlation with Target')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
from scipy.stats import pearsonr



correlations = {}
for feature in x_train.columns:
    corr, _ = pearsonr(x_train[feature], y_train)
    correlations[feature] = abs(corr)

# Converting to a Series for easy plotting
correlation_series = pd.Series(correlations)

# Plotting the histogram
plt.figure(figsize=(10, 6))
correlation_series.plot(kind='bar', color='skyblue')
plt.title('Histogram of Pearson Correlation of Features with Target')
plt.xlabel('Features')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

# Obtention des importances des features
feature_importances = rf.feature_importances_

# Tri des features par leur importance
sorted_idx = np.argsort(feature_importances)[::-1]

# SÃ©lection des features importantes
selected_features_rf = [x_train.columns[i] for i in sorted_idx]
importances_sorted = feature_importances[sorted_idx]

# Affichage des features sÃ©lectionnÃ©es et de leurs importances
selected_features_rf, importances_sorted
```
