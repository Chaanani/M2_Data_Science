```python
import numpy as np

# Transformation des données en ln
medailles2 = medailles.copy()
medailles2['LNPOP'] = np.log(medailles2['POP'])
medailles2['LNPIB'] = np.log(medailles2['PIB'])
medailles2 = medailles2[(medailles2['ANNEES'] == 88) & 
                        (medailles2['TOTALMEDAILLE'].notna()) & 
                        (medailles2['POP'].notna())]
medailles2 = medailles2[['TOTALMEDAILLE', 'LNPOP', 'LNPIB', 'POP', 'PIB', 'PAYS']]

# Affichage des informations du DataFrame transformé
print("Fichier médailles transformées")
medailles2.info()
```

```python
import statsmodels.api as sm

# Estimation : Modèle de comptage (Poisson)
X = new88[['LNPOP', 'LNPIB']]
y = new88['TOTALMEDAILLE']
X = sm.add_constant(X)  # Ajout d'une constante

model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
```

```python
print(model.summary())
```

```python
# Génération des prédictions
new88['XBETA'] = model.predict(X)
new88['PHAT'] = model.mu

```

```python
 #Calcul des probabilités pour chaque observation
# Pour une distribution de Poisson, la probabilité d'observer le nombre exact de 'totalmedaille'
# Calcul de la probabilité de Poisson pour chaque observation
# P(X = k) = (λ^k * e^(-λ)) / k!
new88['PROB'] = new88.apply(lambda row: 
                            (np.exp(-row['PHAT']) * (row['PHAT']**row['TOTALMEDAILLE'])) 
                            / np.math.factorial(int(row['TOTALMEDAILLE'])), axis=1)
# Affichage des résultats
new88[['XBETA', 'PHAT', 'PROB']]
```

```python
import math

# Calcul de la probabilité pour la GB
a_gb = 2
factor_gb = math.factorial(a_gb)
proba_gb = (9.8301**a_gb * math.exp(-9.8301)) / factor_gb
print(f"Probabilité pour la GB: {proba_gb}")
```

```python
# Calcul de la probabilité pour la Russie
a_russia = 142
factor_russia = math.factorial(a_russia)
proba_russia = (39.3968**a_russia * math.exp(-39.3968)) / factor_russia
print(f"Probabilité pour la Russie: {proba_russia}")
```

```python
# Calcul de la probabilité pour le Rawnda
a_rawnda = 0
factor_rawnda = math.factorial(a_rawnda)
proba_rawnda = (.5023**a_rawnda * math.exp(-.5023)) / factor_rawnda
print(f"Probabilité pour le Rawnda: {proba_rawnda}")
```
