```python
import numpy as np
import pandas as pd
from umap import UMAP
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
```

```python
data = pd.read_csv("pasture.csv", sep=" ",header=None)
Y = data.iloc[:, -1]
X = data.iloc[:, : data.shape[1]-1]
X =np.array(X)
Y = np.array(Y)
```

```python
def umapp(neighbors, distance):
    model = UMAP(n_neighbors = neighbors, min_dist = distance, n_components = 2, verbose = True, metric='euclidean', output_metric='euclidean')
    umap = model.fit_transform(X)
    plt.scatter(umap[:, 0], umap[:, 1], c = Y.astype(int))
    plt.title(f"neighbors  {neighbors} and distance {distance}")
    plt.xlabel("UMAP1", fontsize = 20)
    plt.ylabel("UMAP2", fontsize = 20)
    plt.show()
for n in (2, 5, 10, 20, 50, 100, 200):
    for j in [.05*r for r in range(1, 10)]:
        umapp(n, j)
```
